```python
import os
import shutil
import pandas as pd
import numpy as np
```

```python
file_path = "C:\\All data\\IITG\\Projects\\Video Captioning Attempts\\Attempt 1\\Datasets\\MSVD Dataset\\Captions.txt"
```

```python
file1 = open(file_path,'r')
Lines = file1.readlines()
```

```python
df = pd.DataFrame(Lines)
```

```python
df = df.iloc[7:]
df.head()
```

```python
df.columns = ['caption+ID']
df.head()
```

```python
df[['ID', 'captions']] = df['caption+ID'].str.split(' ', 1, expand=True)
```

```python
df.head()
```

```python
df = df.drop(['caption+ID'],axis=1)
df.head()
```

```python
testing_id_path  = "C:\\All data\\IITG\\Projects\\Video Captioning Attempts\\Attempt 1\\Datasets\\MSVD Dataset\\testing_id.txt"
```

```python
file1= open(testing_id_path,'r')
Lines = file1.readlines()
```

```python
dummy = Lines[0][:-5]
```

```python
#df[df['ID']==dummy]
```

```python
testing_df = pd.DataFrame(columns = ['ID','captions'])
testing_df.head()
```

```python
for line in Lines:
    line = line[:-5]
    testing_df = pd.concat([testing_df,df[df['ID']==line]])
    df.drop(df[df['ID'] == line].index, inplace = True)
testing_df.head()
```

```python
video_path = "C:\\All data\\IITG\\Projects\\Video Captioning Attempts\\Attempt 1\\Datasets\\MSVD Dataset\\YouTubeClips\\"
```

```python
training_vids_list = df['ID'].unique()
testing_vids_list = testing_df['ID'].unique()

testing_vids_list
```

```python
images = [f for f in os.listdir() if '.jpg' in f.lower()]

os.mkdir('downloaded_images')

for image in images:
    new_path = 'downloaded_images/' + image
    shutil.move(image, new_path)
```

```python
os.listdir(video_path)
```

```python
videos = [video for video in os.listdir(video_path)]
train_path = "C:\\All data\\IITG\\Projects\\Video Captioning Attempts\\Attempt 1\\Datasets\\MSVD Dataset\\Training Videos\\"
test_path = "C:\\All data\\IITG\\Projects\\Video Captioning Attempts\\Attempt 1\\Datasets\\MSVD Dataset\\Testing Videos\\"

for video in videos:
    if video[:-4] in testing_vids_list:
        shutil.move(video_path+video,test_path+video)
    elif video[:-4] in training_vids_list:
        shutil.move(video_path+video,train_path+video)
```
